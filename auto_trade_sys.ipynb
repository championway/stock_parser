### 勝率分析

```python
out_path = r'D:\Stock\out'
out_list = os.listdir(out_path)
ror_sum = 0
total_num = 0
tp_num = 0
sl_num = 0
for data_out in out_list:
    df_out = pd.read_csv(os.path.join(out_path, data_out))  
    total_num = total_num + len(df_out)
    tp_num = tp_num + df_out[df_out['Status'] == 1].shape[0]
    sl_num = sl_num + df_out[df_out['Status'] == 2].shape[0]
    ror_sum = ror_sum + df_out["RoR"].sum()
    
ror_avg = ror_sum / total_num
ror_day_avg = ror_sum / len(out_list)

print("RoR avg: ", ror_avg)
print("RoR day avg: ", ror_day_avg)
print("RoR sum: ", ror_sum)
print("TP num: ", tp_num)
print("SL num: ", sl_num)
```
